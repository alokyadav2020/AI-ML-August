## Visualizing statstical relationships

```python
tips= sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_theme()
sns.relplot(data=tips,x=tips['total_bill'],y=tips['tip'])
```

```python
sns.relplot(data=tips,x=tips['total_bill'],y=tips['tip'],hue=tips['smoker'])
```

```python
sns.relplot(data=tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

```python
sns.relplot(data=tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

```python
sns.relplot(data=tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['time'])
```

```python
tips['time'].value_counts()
```

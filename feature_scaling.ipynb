## Standardization

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
total_bill = list(df['total_bill'])
total_bill
```

```python
sns.histplot(total_bill)
```

```python
import numpy as np

mean = np.mean(total_bill)

std = np.std(total_bill)
```

```python
mean, std
```

```python
normalisez_data = []
for i in total_bill:
    z_score = (i -mean)/std ## z_score
    normalisez_data.append(z_score)

print(normalisez_data)    
```

```python
normalisez_data
```

```python
sns.histplot(normalisez_data)
```
